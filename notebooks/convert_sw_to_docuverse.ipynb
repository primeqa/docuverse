```python
import json
from docuverse.engines.search_result import SearchResult
from docuverse.engines.search_queries import SearchQueries
```

```python
data=json.load(open('../benchmark/ibmsw/all_searches_apr4_29_52_redo1_t100_r100.json','r'))
```

```python
res = []
qs = []
for i, d in enumerate(data):
    vals = {'text': d['question'],
            'id': str(i),
            'relevant': [d['expected_link']],
            'metadata': {'gold_urls': [d['expected_link']], 'dcc': [], 'gold_docids': []}
        }
    q=SearchQueries.Query(**vals)
    qs.append(q)
    # qs += q
    r = SearchResult(question=q, data=d['documents'])
    res.append(r)
    
with open("../benchmark/ibmsw/queries.jsonl","w") as f:
    for q in qs:
        f.write(q.as_json()+"\n")
        
with open("../output/ibmsw_result.json","w") as f:
    out = [r.as_list() for r in res]
    json.dump(out, f, indent=2)
```
